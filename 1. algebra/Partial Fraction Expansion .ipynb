### Types of Numbers 

This image below is pretty self-explanatory, except for the **Irrational Numbers**. What are irrational numbers? Well, the difference between rational numbers and irrational numbers is that rational numbers are those numbers that are integers and can be expressed in the form of x/y where both numerator and denominator are integers whereas irrational numbers are those numbers which cannot be expressed in a fraction. 

**Irrational Numbers** 

A number is considered as an irrational number if it cannot be simplified further to any fraction of a natural number and an integer. **Side Note (Whole numbers are also called natural numbers.)** Another important thing to note is that the decimal expansion of irrational numbers is neither finite nor recurring. 

### Polynomials 

The word polynomial is derived from the Greek words 'poly' which means 'many' and 'nominal' which means 'terms'. So altogether it means 'many terms'. A polynomial can have any number of terms but not infinite. 

Polynomial is made up of at least two terms. A polynomial is defined as an expression which is composed of variables, coefficients, constants and exponents, that are combined using the mathematical operations such as addition, subtraction, multiplication and division (No division operation by a variable). 

Bsed on the number of terms present in the expression, it is classified as **monomial, binomial, and trinomial.**  

The **degree of a polynomial** is defined as the highest degree of a monomial within a polynomial.

### Rational Expression 

A rational expression (AKA rational fraction) is simply a fraction whose numerator and denominator are polynomials. 

### Partial Fraction (of rational expression) 

Partial fraction is each of two or more fractions into which a more complex fraction can be decomposed as a sum. 

Example: 

$\frac{5x-4}{x^2-x-2} = \frac{2}{x-2} + \frac{3}{x+1}$

It is easy to go from the right-hand side to the left-hand side. But there are cases where we need to make a rational expression into partial fractions. So how do we do that? 

**Side Note: Factorization** 
In mathematics, factorization or factoring consists of writing a number or another mathematical object as a product of several factors. 

First, we look at the denominator of left-hand side. It is a 2nd degree polynomial, a quadratic polynomial. We would like do the factorization. We can use the **quadratic formula** to do the factorization. (But not all the time). 

$X = \frac{-b \ \pm \ \sqrt{b^2 \ - \ 4ac}}{2a}$

Derivation of quadratic formula starts from this equation: $ax^2 + bx + c = 0$. This equation is equation is true under several premises. Positive $a$ will make the graph upward parbola. But if $c$ is positive, then the parabola will never cross the x-axis. 
So in this case, there will be no solutions. 

In this case, we have to use the method of **completing the square**. 

We know the following is true: 
$(x+\lambda)^2 = x^2 + 2\lambda x + \lambda^2$

And say we are given this equation. 
$ax^2 + bx + c$

We have to factor out $a$ from the equation and complete the square to make it look like $x^2 + 2\lambda x + \lambda^2$


$a(x^2 + \frac{b}{a}x + \frac{c}{a}) \Longrightarrow a(x^2 + \frac{b}{a}x + \alpha + \frac{c}{a} - \alpha)$

$2\lambda = \frac{b}{a} \ \ $ and $ \ \ \lambda^2 = \alpha$

So we get $a((x + \lambda)^2 + \frac{c}{a} - \alpha)$

**Let's come back to our example** 

Okay, so the denominator of our example satisfies all the requirements to use the quadratic equation. But I want to think about it for a moment. What is the logic behind the qudratic equation? Well, like I said before, the derivation of the quadratic formula starts from $ax^2 + bx + c = 0$. And if we solve for $x$, then we get the quadratic formula. So whatever value we get for $x$ makes the equation $0$. And most of the times, we will get two different values for $x$, because there is a plus-minus sign in the quadratic formula. And let's just pick one of the two answers. 

For our example, we get $x = 2$ and $x = -1$. 

I'll pick $x = 2$. Then the following equation, $(x-2)=0$ satisfies. And this expression, $(x-2)$ is a factor of $(x^2-x-2)$. 

And interestingly enough, if you divide $(x^2-x-2)$ by $(x-2)$, we get $(x+1)$.

Therefore, $(x^2-x-2) = (x-2)(x+1)$


Okay, now we have done the dirty work and we have the following equation. 

$\frac{5x-4}{(x-2)(x+1)}$

And we would like to expand this equation into two partial fractions. This process is called **partial fraction expansion**. 

$\frac{5x-4}{(x-2)(x+1)} = \frac{A}{(x-2)} + \frac{B}{(x+1)}$

So we have to figure out what $A$ and $B$ are. 

$\frac{A}{(x-2)} + \frac{B}{(x+1)} = \frac{A(x+1) + B(x-2)}{(x-2)(x+1)}$

Therefore,

$5x-4 = A(x+1) + B(x-2)$

If $x = -1$, then $-9 = B(-3)$. Therefore, $B = 3$

If $x = 2$, then $6 = A(3)$. Therefore, $A = 2$

Finally, $\frac{5x-4}{(x-2)(x+1)} = \frac{2}{(x-2)} + \frac{3}{(x+1)}$
