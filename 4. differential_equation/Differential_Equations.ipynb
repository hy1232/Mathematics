#### Quick Summary of what I have studied (Differential Equations for Dummies) 

What is differential equation? **Differential equation is an equation involving derivatives of a function or functions.** 

What is the definition of derivative? It is the rate of change. Derivative answers how much the output variable changes when the input variable changes by little. To formulate that mathematically, we have the following equation: 

$\frac{dy}{dx} = \lim\limits_{h \to 0} \frac{y(x + h) - y(x)}{h}$

So say we have $y(x) = x^2$

Then, $\frac{dy}{dx} = \lim\limits_{h \to 0} \frac{(x + h)^2 - x^2}{h}$

If we solve it, we will get $\frac{dy}{dx} = 2x + h$ and $h$ is infinitesimal. But calculating this is such a hassle. So, alternatively we can just take the derivative of the function, $y(x)$ and we will arrive at the same answer. 

But what we do in **Differential Equation** is making different alterations to the differential equation to acheive its original equation, $y(x)$. And we simply call this process solving for $y(x)$.

* Classifying linear versus nonlinear equations 

You call a differential equation linear if it exclusively involves linear terms (that is, terms to the power of 1) of y. 

Examples: 

$1. \ \frac{dy}{dx} = 5$ $\;\;\;\;\;\;$ $2. \ \frac{dy}{dx} = g(x)$ $\;\;\;\;\;\;$ $3. \ \frac{dy}{dx} = y + 1$ $\;\;\;\;\;\;$ $4. \ \frac{dy}{dx} = 3y + 1$ $\;\;\;\;\;\;$ $5. \ \frac{dy}{dx} = g(x)y$ $\;\;\;\;\;\;$ $6. \ \frac{dy}{dx} = g(x)y + q(x)$

These are different kinds of first order linear differential equations. And different kinds of methods are involved to solve them. 

**Disclaimer:** One type of equation doesn't always involve one type of method to solve it, meaning there could be many different techniques to solve one type of equation. In other words, one technique could be used to solve two different types of equations. So choosing what techniques to use for a given equation is the matter of personal preference. But for every type of differential equation, there is always a easiest way and we tend to choose the easiest.  

**Generally,** simple equations like the first two equations given above as examples could be solved by taking integral w.r.t. $x$ on both sides. But the rest of the equations won't be solved that way. That is why we learn different techniques to solve them and prove why those different techniques work. 

**Interesting Point:** Say I have an equation, $\frac{dy}{dt} = t^3 - t^2$. Well, this looks exactly like the example equation number 2. #### First Order Linear Differential Equations 

First order linear differential equation has the following form: 

$\frac{dy}{dx} + P(x) \cdot y = Q(x)$

And then I have the integrating factor. 

$I(x) = e^{\int P(x) dx}$

Once you have the integrating factor, you can now write the general solution to the differential equation. 

$y = \frac{1}{I(x)}[\int I(x) \cdot Q(x) \ dx + C]$

**Example** 

Let's say I have the following differential equation: 

$y' + 2y = 2e^x$

$\begin{cases}
P(x) = 2 \\ 
Q(x) = 2e^x
\end{cases}$

$I(x) = e^{\int P(x) dx} = e^{\int 2 dx} = e^{2x}$

Therefore, 

$y = \frac{1}{e^{2x}} [\int e^{2x} \cdot 2e^x dx + C]$

$y = \frac{1}{e^{2x}} [2 \int e^{3x} dx + C]$

$y = \frac{1}{e^{2x}} [2 \cdot \frac{e^{3x}}{3} + C]$

And this is the final answer: 

$y = \frac{2}{3} \cdot e^{x} + C \cdot e^{-2x}$

$I(x)$ is called the **integrating factor**.

**Definition of Integrating factor:** In mathematics, an integrating factor is a function that is chosedn to facilitate the solving of a given equation involving differentials. 

#### Proof for the Integrating Factor, $I(X) = e^{\int P(x)dx}$

We start with this form: 

$\frac{dy}{dx} + P(x) \cdot y = Q(x)$

Multiply $I(x)$ on both sides. 

$I(x) \cdot \frac{dy}{dx} + I(x) \cdot P(x) \cdot y = I(x) \cdot Q(x)$

##### I define $y = u$ and $I(x) = v$

$v \cdot \frac{du}{dx} + v \cdot P(x) \cdot u = v \cdot Q(x)$

##### I also define that $v \cdot P(x) = \frac{dv}{dx}$ This is critical!

So I will have the following: 

$v \cdot \frac{du}{dx} + u \cdot \frac{dv}{dx} = v \cdot Q(x)$

So back to $v \cdot P(x) = \frac{dv}{dx}$. For this to be true, what should $v$ be? 

So we have to solve for $v$ using separation of variables, since it is separable differential equation. 

$P(x) \cdot dx = \frac{1}{v} \cdot dv$

$\int P(x) \cdot dx = \int \frac{1}{v} \cdot dv$

$ln(v) = \int P(x) \cdot dx$

$v = e^{\int P(x) \cdot dx}$

For $v \cdot P(x) = \frac{dv}{dx}$ to be true, $v = e^{\int P(x) \cdot dx}$ has to be true. 

And we actually defined $I(x) = v$

So $I(x) = e^{\int P(x) \cdot dx}$


#### Proof for the $Y = \frac{\int I(x) \cdot Q(x) \ dx + C}{I(X)}$

The proof for $I(x)$ is not enough to know why $Y$ is such. 

$v \cdot \frac{du}{dx} + u \cdot \frac{dv}{dx} = v \cdot Q(x)$

From here, we can use the product rule backwards: $\frac{df(x)g(x)}{dx} = f'(x)g(x) + g'(x)f(x)$

$\frac{d(v \cdot u)}{dx} = v \cdot Q(x)$

Now, we would like to solve for $u$ which is $y$. So we take the integral w.r.t. $x$ on both sides. 

$\int \frac{d(v \cdot u)}{dx} \cdot dx = \int v \cdot Q(x) \cdot dx$

From here, I used **integration by parts** technique to get the next equation. (Well, I didn't really use it. I just got the idea how to go to the next step from this.) 

The integral of the derivative of $(u \cdot v)$ is just $(u \cdot v)$.

$v \cdot u = \int v \cdot Q(x) \cdot dx$

$u = \frac{1}{v}\int v \cdot Q(x) \cdot dx$

$y = \frac{1}{I(x)}\int I(x) \cdot Q(x) \cdot dx$

**Turning nonlinear separable into linear separable differential equation** (This is sometimes referred to as **First Order Homogeneous Differential Equation**)

Not every nonlinear differential equations are separable, just like not every linear differential equations are separable. But there are nonlinear differential equations that are seemingly nonseparable at first glace but are actually separable. Take a look at the following: 

$\frac{dy}{dx} = \frac{2y^4 + x^4}{xy^3}$

Does this look separable? No! But we still have to check if it is actually separable. The way to check is to see if the following conditions are satisfied for the given differential equation. 

$\frac{dy}{dx} = f(x,y)$

when $f(x,y) = f(tx, ty)$, where $t$ is a constant. 

For our given equation, this condition is satisfied. $\therefore$ it is separable. But how do we do it? **We can use the trick of setting $y = xv$.**

$\frac{d(xv)}{dx} = \frac{2(xv)^4+x^4}{x(xv)^3}$

$\frac{d(xv)}{dx} = \frac{(2v^4+1)}{v^3}$

$\frac{d(y)}{dx} = \frac{(2v^4+1)}{v^3}$

The term $y$ went away on the right hand side. So maybe in that sense, it has been transformed from non-linear into linear.

$\frac{d(xv)}{dx} = \frac{(2v^4+1)}{v^3}$

$v + x \frac{dv}{dx} = \frac{(2v^4+1)}{v^3}$

At the end, the answer we get is $y = (x^4m^4-1)^{1/4} \cdot x$

$\frac{dy}{dx} = \frac{2y^4+x^4}{xy^3}$

We can rearrange them and express them as the following: 

$-(2y^4+x^4)+xy^3 \frac{dy}{dx} = 0$

$M(x,y) + N(x,y) \frac{dy}{dx} = 0$

Before we go any further, I would like to mention one concept. 

If a function $f(x,y)$ satisfies the following condition, then it is called **homogeneous function**. 

$f(tx,ty) = t^kf(x,y)$

And $k$ is called the degree of homogeneity.

So if $M(x,y)$ and $N(x,y)$ are homogeneous functions with the same degree of homogeneity, then the following is also true. 

$\frac{dy}{dx} = f(x,y)$

$f(x,y) = f(tx, ty)$

And if this condition holds true, then we call this differential equation **First Order Homogeneous Differential Equation**. This is just another way of explaining what I described above. If it is a homogeneous differential equation, then we have to transform the equation, so that it can be separable. And that trick is described above.

An Exact Equation is just another form of differential equation and it is followed by a certain technique to solve it. 

Let's take a look at this: 

$M(x,y) + N(x,y) \cdot \frac{dy}{dx} = 0$

This is a general form of exact differential equation, but we can never be sure that it is an exact equation unless it satisfies a certain condition which is 

$\frac{\partial M(x,y)}{\partial y} = \frac{\partial N(x,y)}{\partial x}$. 

Why does such condition have to hold true and what does that even mean? To answer these questions, we have to think about this multivariable function, $f(x,y)$. If we were to set this function equal to some constant, then we would express it as 

$f(x,y) = C$ 

When we are setting the function equal to some constant, then the dimension of the function goes from 3 to 2. And now we can think of $y$ as a function of $x$. 

Now, let's take a derivative wrt $x$ on both sides. Taking derivative wrt x instead of partial already means that we are treating this function as an univariate function. 

$\frac{d f(x,y)}{dx} = 0$

If we were to expand it, 

$\frac{d f(x,y)}{dx} = \frac{\partial f(x,y)}{\partial x} + \frac{\partial f(x,y)}{\partial y} \cdot \frac{dy}{dx} = 0$
**(There is a side note on this expansion)** 

Partial $f(x,y)$ wrt x is referred to as $M(x,y)$ and partial $f(x,y)$ wrt y is referred to as $N(x,y)$.

$\frac{\partial f(x,y)}{\partial x} = M(x,y)$

$\frac{\partial f(x,y)}{\partial y} = N(x,y)$

And as we mentioned before the condition, $\frac{\partial M(x,y)}{\partial y} = \frac{\partial N(x,y)}{\partial x}$ 
has to be true. This is same thing as $\frac{\partial^2 f(x,y)}{\partial y \partial x} = \frac{\partial^2 f(x,y)}{\partial x \partial y}$. 
 
Now, we are ready to answer the question we asked previously. If $f(x,y)$ is a continuous function, then such condition is true because of the following. 

By definition, a derivative of a function is the following: 

$\frac{df(x)}{dx} = \lim\limits_{h \to 0} \frac{f(x + h) - f(x)}{h}$

Then the following is also true. 

$\frac{\partial f(x,y)}{\partial x} = \lim\limits_{h \to 0} \frac{f(x + h, \ y) - f(x,y)}{h}$

and 

$\frac{\partial^2 f(x,y)}{\partial y \partial x} = \lim\limits_{h \to 0 \ k \to 0} \frac{f(x + h, \ y + k) - f(x,y)}{hk}$

But this is same as $\frac{\partial^2 f(x,y)}{\partial x \partial y}$ as you probably have guessed already. In other words, the direction of the change and the end point is the same. It is just a matter of which direction you go first. But whether you go along the x-axis first or the y-axis first doesn't matter.  

This condition is true only if there is a continuous function $f(x,y)$. So we can use that condition to see if there is a continuous function $f(x,y)$. If so, then we can be sure of the existence of the function, $f(x,y)$. 

And if we set this equation equal to some constant $C$, then the expression would be like the following: 

$f(x,y) = C$ 

And if we take derivative wrt $x$ on both sides, then we get 

$\frac{df(x,y)}{dx} = \frac{\partial f(x,y)}{\partial x} + \frac{\partial f(x,y)}{\partial y} \cdot \frac{dy}{dx} = 0$

But normally, when we face a differential equation problem, we will be given a equation like the following: 

$\frac{\partial f(x,y)}{\partial x} + \frac{\partial f(x,y)}{\partial y} \cdot \frac{dy}{dx} = 0$

And from here, our goal is the get $f(x,y)$ from the different equation. The next block of latex code is the explanation for how to acheive $f(x,y)$ from the differential equation.

Say I have $f(x,y) = g(x) \cdot h(y) + a(x) + b(y) = C$

Of course $y = f(x)$. 

And if we take derivative wrt $x$, 

$\frac{df(x,y)}{dx} = \frac{\partial f(x,y)}{\partial x} + \frac{\partial f(x,y)}{\partial y} \cdot \frac{dy}{dx} = 0$

$ = \frac{\partial [g(x)h(y) + a(x) + b(y)]}{\partial x} + \frac{\partial [g(x)h(y) + a(x) + b(y)]}{\partial y} \cdot \frac{dy}{dx} = 0$

$ = [g'(x) \cdot h(y) + a'(x)] + [g(x) \cdot h'(y) + b'(y)] \cdot \frac{dy}{dx} = 0$

Take $[g'(x) \cdot h(y) + a'(x)]$ first and take the integral wrt $x$. 

$\int [g'(x) \cdot h(y) + a'(x)] dx = g(x) \cdot h(y) + a(x)$

Well, $g(x) \cdot h(y) + a(x)$ is insufficient to say it is $f(x,y)$. Because during the partial derivative wrt $x$ process, $b(y)$ has been removed. Therefore, we have to see the next part of the differential equation. 

Take $[g(x) \cdot h'(y) + b'(y)] \cdot \frac{dy}{dx}$ and take the integral wrt $x$.

$\int [g(x) \cdot h'(y) + b'(y)] \cdot \frac{dy}{dx} \cdot dx$

$= \int [g(x) \cdot h'(y) + b'(y)] \cdot dy$

$= g(x) \cdot h(y) + b(y)$

So we have two answers, $g(x) \cdot h(y) + a(x)$ and $g(x) \cdot h(y) + b(y)$. And we can compare and see what is missing for both sides. 

If we conclude that the answer is $g(x) \cdot h(y) + a(x) + b(y)$ then there we have the function $f(x,y)$. 

And if we set it equal to some constant, then we have $g(x) \cdot h(y) + a(x) + b(y) = C$. If it is possible, we can solve for $y$ and get the explicit solution. If not, we can just leave it as implicit solution.

And $b^2 - 4ac = 0$. 

$\therefore r = \frac{-b}{2a}$

There is only one option for $r$. 

So $y = C \cdot e^{\frac{-bx}{2a}}$

However, say we are given two initial conditions: $y(0) = 1$ and $y'(0) = 2$. Then $y = C \cdot e^{\frac{-bx}{2a}}$ is not general enough to satisfy these two initial conditions. Therefore, we need something more general. 

So we need an equation like the following: $y = m(x) \cdot e^{\frac{-bx}{2a}}$. This is equation is more general. Now our job is to figure out what $m(x)$ is. 

This equation will satisfy $ay'' + by' + cy = 0$. 

$y = m(x) \cdot e^{\frac{-bx}{2a}}$

$y' = m'(x) \cdot e^{(\frac{-bx}{2a})} + m(x) \cdot (\frac{-b}{2a}) \cdot e^{\frac{-bx}{2a}}$ 

$y'' = e^{\frac{-bx}{2a}} \cdot [m''(x) + (-\frac{b}{a}) \cdot m'(x) + m(x) \cdot (\frac{b}{2a})^2]$

If we plug in these equations into the differential equation, $ay'' + by' + cy = 0$, we get 

$m''(x) = 0$

In the middle of this process, we have to assume the fact that $c - \frac{b^2}{4a} = 0$ because the discriminant is equal to zero. $b^2 - 4ac = 0$

Then $\int \int m''(x)dxdx = m(x) = d_1x + d_2$

$\therefore y = (d_1x + d_2) \cdot e^{\frac{-bx}{2a}}$

**Chapter 6** 

Nonhomogeneous Differential Equations (Second Order Linear) 

Say I have the following differential equation: 

$Ay'' + By' + Cy = g(x)$

And $h(x)$ is a solution for the homogeneous differential equation, $Ay'' + By' + Cy = 0$.

So $Ah''(x) + Bh'(x) + Ch(x) = 0$

And $j(x)$ is a partial solution for $Ay'' + By' + Cy = g(x)$. 

Then $h(x) + j(x)$ is the general solution for $Ay'' + By' + Cy = g(x)$. 

$A(h''(x) + j''(x)) + B(h'(x) + j'(x)) + C(h(x) + j(x)) = g(x)$ 

$[Ah''(x) + Bh'(x) + Ch(x)] + [Aj''(x) + Bj'(x) + Cj(x)] = g(x)$

The first part of the left hand side corresponds to zero, and the right part of the left hand side corresponds to $g(x)$. 

**Let's see an example** 

$y'' - 3y' - 4y = 3e^{2x}$

The solution for the homogeneous equation is the following: 

$r = 4$ and $r = -1$

$y_h = C_1e^{4x} + C_2e^{-x}$

Now, we need the solution for the $g(x)$ part, which is $3e^{2x}$.

**Method of Undetermined Coefficients** 

For whatever the $g(x)$ is, we kind of have to guess what the solution would be. In this case, we guess that the partial solution is some kind of function that looks similar to $g(x)$. 

$y_p = K \cdot e^{2x}$

Here, we used $K$ as a constant. But this constant is not determined. That is why this is called undetermined coefficient. We plug this solution into the differential equation, and then we figure out what the coefficient of the solution is. 

$
\begin{cases}
y_p = K \cdot e^{2x} \\
y'_p = 2K \cdot e^{2x} \\ 
y''_p = 4K \cdot e^{2x} 
\end{cases} 
$

$4K \cdot e^{2x}  - 3(2K \cdot e^{2x}) -4(K \cdot e^{2x}) = 3e^{2x}$

$\therefore K = -\frac{1}{2}$

$y_p = -\frac{1}{2}e^{2x}$

So, putting together the solution for the homogeneous differential equation and the partial solution for $g(x)$, we get: 

$y = C_1e^{4x} + C_2e^{-x} - \frac{1}{2}e^{2x}$

I won't solve for $C_1$ and $C_2$. But one thing I want to take note of is the following. Say I have two initial conditions. For these conditions, the constants, $C_1$ and $C_2$ will be different from $C_1$ and $C_2$ of the homogeneous differential equation with the same given initial conditions. That means I shouldn't determine the constants for the homogeneous differential equation when we are dealing with non-homogeneous differential equation. I should determine the constants for the non-homogeneous differential equations with the initial conditions, not for the homogeneous differential equation. 

**Question** 

Why is the solution for the homogeneous differential equation and the partial solution for $g(x)$ is the general solution for the non-homogeneous differential equation? 

**Answer:** We would like to get a solution in the most general form, which could represent all possible answers for different initial conditions. That is why we add the solution for the homogeneous differential equation. The solution for the homogeneous differential equation has two constants and these allows me to modify the solution in a way that it satifies the initial conditions. 

**Another example** 

$y'' - 3y' - 4y = 2sin(x)$

I have only altered the $g(x)$ for the simplicity. 

$y_h = C_1e^{4x} + C_2e^{-x}$

Again, I have to guess what the partial solution would be. 

$y_p = A \cdot sin(x) + B \cdot cos(x)$ 

I have guess the solution like above, because $g(x)$ is a function of $sin(x)$, and the derivative and the second derivative of $cos(x)$ are $-sin(x)$. 

$
\begin{cases}
y_p = A \cdot sin(x) + B \cdot cos(x) \\ 
y'_p = A \cdot cos(x) - B \cdot sin(x) \\ 
y''_p = -A \cdot sin(x) - B \cdot cos(x) 
\end{cases}
$

If I plug them into the differential equation and simplify them, I get 

$sin(x)[3B - 5A] + cos(x)[-3A - 5B] = 2sin(x)$

Now, I have to determine what $A$ and $B$ are. 

$
\begin{cases} 
3B - 5A = 2  \\
-3A - 5B = 0
\end{cases} 
$

$A = - \frac{5}{17}$ and $B = \frac{3}{17}$

$\therefore y = C_1e^{4x} + C_2e^{-x} + (-\frac{5}{17})sin(x) + (\frac{3}{17})cos(x)$
